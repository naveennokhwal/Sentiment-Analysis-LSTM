### Distribution of labels

```python
label_counts = data['sentiment'].value_counts()
```

```python
# Plotting the distribution of labels
plt.figure(figsize=(4,3))
label_counts.plot(kind='bar', color=['blue', 'red', 'green'])
plt.title('Distribution of Sentiment Labels')
plt.xlabel('Sentiment Labels')
plt.ylabel('Count')
plt.xticks(rotation=0)

# Plotting the distribution of labels as a pie chart
plt.figure(figsize=(4,4))
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sentiment Labels')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### Text Length Analysis

```python
# calculating numbers of characters in each doucmemt
char_length = data['text'].apply(len)

# calculating numbers of words in each doucmemt
word_count = data['text'].apply(lambda x: len(x.split()))
```

```python
# Plotting histogram of text lengths (word count)
plt.figure(figsize=(4, 3))
plt.hist(word_count, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Text Lengths (Word Count)')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```
